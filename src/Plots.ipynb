```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
recall_linear_swil = np.zeros((9,15))
for i in range(5):
    loaded = np.load('./logs/linear_fashionmnist_holdout_[8]_swil_recall' + str(i) + '.npy')
    #print(recall_linear_swil)
    #print(loaded)
    for k in range(len(loaded)):
        for j in range(len(loaded[k])):
            recall_linear_swil[k][j] += loaded[k][j]
            
#print(loaded[0])
recall_linear_swil /= 5
```

```python
for c in recall_linear_swil:
    plt.plot(c)

plt.legend(list(range(8)) + [9])
plt.title("Linear FMNIST Recalls per Class")
plt.xlabel("Epochs")
plt.ylabel("Recall")
plt.show()
```

```python
similar_old = (recall_linear_swil[5] + recall_linear_swil[7])/2


plt.plot(similar_old)
plt.title("Linear FMNIST SWIL Recall Similar Old Classes")
plt.xlabel("Epochs")
plt.ylabel("Recall")
plt.xlim([0, 5])
plt.ylim([0.6, 1])
plt.show()
```

```python
similar_old = (recall_linear_swil[0] + recall_linear_swil[1] + recall_linear_swil[2] + recall_linear_swil[3] + recall_linear_swil[4] + recall_linear_swil[6])/6


plt.plot(similar_old)
plt.title("Linear FMNIST SWIL Recall Different Old Classes")
plt.xlabel("Epochs")
plt.ylabel("Recall")
plt.xlim([0, 5])
plt.ylim([0.6, 1])
plt.show()

# NOTE: recalls are lower and this is probably because our general accuracy is a little lower as well
```
